Import de Dados

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

with open('bfs_stats.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data)
df.head()
```

Gráfico Custo computacional (tempo × nós visitados)

```python
plt.figure(figsize=(8,5))
plt.scatter(df['nodesVisited'], df['timeMs'], alpha=0.7, color='royalblue')
plt.title("Custo Computacional: Tempo x Nós Visitados")
plt.xlabel("Nós Visitados")
plt.ylabel("Tempo de Execução (ms)")
plt.grid(True)
plt.show()
```

Gráfico Crescimento com aumento da distância (pathLength × tempo)

```python
plt.figure(figsize=(8,5))
plt.plot(df['pathLength'], df['timeMs'], 'o-', color='seagreen')
plt.title("Crescimento com o Tamanho do Caminho")
plt.xlabel("Comprimento do Caminho (nós)")
plt.ylabel("Tempo (ms)")
plt.grid(True)
plt.show()
```

Grafico Relação entre memória usada e tempo (maxQueueSize × timeMs)

```python
plt.figure(figsize=(8,5))
plt.scatter(df['maxQueueSize'], df['timeMs'], alpha=0.7, color='darkorange')
plt.title("Uso de Memória x Tempo de Execução")
plt.xlabel("Tamanho Máximo da Fila (maxQueueSize)")
plt.ylabel("Tempo (ms)")
plt.grid(True)
plt.show()
```

Gráfico Eficiência (tempo por nó)

```python
df['timePerNode'] = df['timeMs'] / df['nodesVisited']

plt.figure(figsize=(8,5))
plt.plot(df.index, df['timePerNode'], color='purple')
plt.title("Eficiência: Tempo Médio por Nó Visitado")
plt.xlabel("Execução (agente)")
plt.ylabel("Tempo por Nó (ms/nó)")
plt.grid(True)
plt.show()
```

Médias finais

```python
print("Tempo médio de execução: ", df['timeMs'].mean(), "ms")
print("Nós visitados em média: ", df['nodesVisited'].mean())
print("Tamanho médio do caminho: ", df['pathLength'].mean())
print("Uso máximo médio da fila: ", df['maxQueueSize'].mean())
```
